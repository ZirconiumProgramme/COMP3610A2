Data Processing and Feature Engineering

```python
#Temporal Features
df = pl.read_parquet('Transformed_TripData.parquet')

df = df.filter((pl.col("payment_type") == 1))

df = df.with_columns([
    pl.col("tpep_pickup_datetime").dt.hour().alias("pickup_hour"),
    (pl.col("tpep_pickup_datetime").dt.weekday() - 1).alias("pickup_day_of_week"),
    pl.col("tpep_pickup_datetime").dt.weekday().is_in([6, 7]).alias("is_weekend").cast(pl.Int8),
    pl.col("trip_distance") / (pl.col("trip_duration_minutes") / 60).alias("trip_speed_mph")
])

#Trip features(Trip Speed MPH was already included)



df = df.with_columns([
   # Added the speed calculation here so it actually saves
   (pl.col("trip_distance") / (pl.col("trip_duration_minutes") / 60)).alias("trip_speed_mph"),
   ((pl.col("tpep_dropoff_datetime") - pl.col("tpep_pickup_datetime")).dt.total_milliseconds() / 60000).alias("trip_duration_minutes"),
   pl.col("trip_distance").log1p().alias("log_trip_distance")
])


#Fare Features

df = df.with_columns([
    pl.when(pl.col("trip_distance") > 0).then(pl.col("fare_amount") / pl.col("trip_distance")).otherwise(0).alias("fare_per_mile"),
    pl.when(pl.col("trip_duration_minutes") > 0).then(pl.col("fare_amount") / pl.col("trip_duration_minutes")).otherwise(0).alias("fare_per_minute")
])

#Zone Features
df = df.to_dummies(["PU_Borough", "DO_Borough"])
```

```python
df = df.with_columns([
   pl.when(pl.col("tip_amount") > pl.col("fare_amount") * .20  )
   .then(1)
   .otherwise(0)
   .alias("high_tip")
])
```

```python
features = df.drop(['tpep_pickup_datetime', 'tpep_dropoff_datetime', 'tip_amount', 'high_tip', 'store_and_fwd_flag', 'trip_duration', 'PU_Zone', 'PU_service_zone', 'DO_Zone', 'DO_service_zone', 'total_amount'])

X = features.to_pandas()
print(X.dtypes)
```

```python

y_1 = df['tip_amount'].to_pandas()
y_2 = df['high_tip'].to_pandas()



print("Number of rows:", len(X))

X_train, X_temp, y_train_1, y_temp_1, y_train_2, y_temp_2 = train_test_split(
    X, y_1, y_2, test_size=0.30, random_state=42, stratify=y_2
)

X_val, X_test, y_val_1, y_test_1, y_val_2, y_test_2 = train_test_split(
    X_temp, y_temp_1, y_temp_2, test_size=0.50, random_state=42, stratify=y_temp_2
)


scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

for name, y in [("Train", y_train_2), ("Val", y_val_2), ("Test", y_test_2)]:
    print(f"{name} split size: {len(y)}")
    print(f"{name} High Tip distribution:\n{y.value_counts(normalize=True)}\n")

print("Features used:")
print(X.dtypes)

print("\nExcluded Features: ['tpep_pickup_datetime', 'tpep_dropoff_datetime', 'tip_amount', 'high_tip', 'store_and_fwd_flag']")
print("Reason: Original timestamps were converted to numerical features; 'tip_amount' and 'high_tip' are targets (leakage prevention); 'store_and_fwd_flag' is a categorical string excluded for simplicity.")
```

Model Training and Tuning 

```python
#4a
linearReg = LinearRegression()
linearReg.fit(X_train_scaled, y_train_1) 
y_pred_linear = linearReg.predict(X_val_scaled)


randomForestRegressor = RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42, n_jobs=-1)
randomForestRegressor.fit(X_train_scaled, y_train_1)
y_pred_rfr = randomForestRegressor.predict(X_val_scaled)


#4b

logisticClassifier = LogisticRegression(max_iter=1000, random_state=42)
logisticClassifier.fit(X_train_scaled, y_train_2)
y_pred_logclf = logisticClassifier.predict(X_val_scaled)
y_prob_logclf = logisticClassifier.predict_proba(X_val_scaled)[:, 1] 


randomForestClassifier = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42, n_jobs=-1)
randomForestClassifier.fit(X_train_scaled, y_train_2) 
y_pred_rfcl = randomForestClassifier.predict(X_val_scaled) 
y_prob_rfcl = randomForestClassifier.predict_proba(X_val_scaled)[:, 1] 
```

```python
# 4c: Report performance on the validation set

def print_regression_metrics(model_name, y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = numpy.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    print(f"--- {model_name} (Regression) ---")
    print(f"MAE: {mae:.4f} | RMSE: {rmse:.4f} | R²: {r2:.4f}\n")

# --- Classification Metrics ---
def print_classification_metrics(model_name, y_true, y_pred, y_prob):
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred)
    rec = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    auc = roc_auc_score(y_true, y_prob)
    print(f"--- {model_name} (Classification) ---")
    print(f"Accuracy: {acc:.4f} | Precision: {prec:.4f} | Recall: {rec:.4f} | F1: {f1:.4f} | AUC-ROC: {auc:.4f}\n")

print("=== VALIDATION SET PERFORMANCE ===\n")

# Evaluate Regression Models (using y_val_1)
print_regression_metrics("Linear Regression", y_val_1, y_pred_linear)
print_regression_metrics("Random Forest Regressor", y_val_1, y_pred_rfr)

# Evaluate Classification Models (using y_val_2)
print_classification_metrics("Logistic Regression", y_val_2, y_pred_logclf, y_prob_logclf)
print_classification_metrics("Random Forest Classifier", y_val_2, y_pred_rfcl, y_prob_rfcl)
```

```python
from sklearn.model_selection import RandomizedSearchCV

# Requirement 5b: Sample the training data to speed up the search (200k-500k rows)
# We use .sample() on the indices to keep the data and labels aligned
sample_idx = numpy.random.choice(len(X_train_scaled), size=300000, replace=False)
X_sample = X_train_scaled[sample_idx]
y_sample = y_train_2.iloc[sample_idx]

# Requirement 5a: Define a search space with at least 3 hyperparameters
param_dist = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Requirement 5a & 5b: Setup RandomizedSearchCV with 5-fold CV
# n_iter=20 as per performance expectations
rf_random_search = RandomizedSearchCV(
    estimator=RandomForestClassifier(random_state=42, n_jobs=-1),
    param_distributions=param_dist,
    n_iter=20, 
    cv=5, 
    random_state=42,
    scoring='f1', # Optimization target
    verbose=1
)

# Run the search
rf_random_search.fit(X_sample, y_sample)

# 5c
print(f"Best Parameters Found: {rf_random_search.best_params_}")

#5d
tuned_rf = rf_random_search.best_estimator_
y_pred_tuned = tuned_rf.predict(X_val_scaled)
y_prob_tuned = tuned_rf.predict_proba(X_val_scaled)[:, 1]

print_classification_metrics("Tuned Random Forest", y_val_2, y_pred_tuned, y_prob_tuned)
```

```python


# Check for Apple Silicon GPU (MPS) acceleration
device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")
print(f"Using device: {device}")

# 1. Prepare Data (Requirement 6b)
# Convert scaled numpy arrays to PyTorch Tensors
X_train_tensor = torch.tensor(X_train_scaled, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train_2.values, dtype=torch.float32).view(-1, 1)
X_val_tensor = torch.tensor(X_val_scaled, dtype=torch.float32)
y_val_tensor = torch.tensor(y_val_2.values, dtype=torch.float32).view(-1, 1)

# Create DataLoader for batch processing
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
train_loader = DataLoader(train_dataset, batch_size=1024, shuffle=True)

# 2. Define Architecture (Requirement 6a)
class TipNeuralNet(nn.Module):
    def __init__(self, input_dim):
        super(TipNeuralNet, self).__init__()
        self.network = nn.Sequential(
            nn.Linear(input_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 32), # Hidden Layer 2
            nn.ReLU(),
            nn.Linear(32, 1)   # Output Layer
        )
        
    def forward(self, x):
        return self.network(x)

# Initialize model, loss, and optimizer
model = TipNeuralNet(X_train_scaled.shape[1]).to(device)
criterion = nn.BCEWithLogitsLoss() # Use for classification 
optimizer = optim.Adam(model.parameters(), lr=0.001)

# 3. Training Loop (Requirement 6b)
train_losses, val_losses = [], []
epochs = 20 # Minimum requirement 

for epoch in range(epochs):
    model.train()
    batch_losses = []
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        batch_losses.append(loss.item())
    
    # Validation phase
    model.eval()
    with torch.no_grad():
        val_outputs = model(X_val_tensor.to(device))
        val_loss = criterion(val_outputs, y_val_tensor.to(device))
        
    train_losses.append(sum(batch_losses)/len(batch_losses))
    val_losses.append(val_loss.item())
    
    if (epoch + 1) % 5 == 0:
        print(f"Epoch {epoch+1}/20 | Train Loss: {train_losses[-1]:.4f} | Val Loss: {val_losses[-1]:.4f}")

# 4. Plot Loss Curves (Requirement 6c)
plt.figure(figsize=(8, 5))
plt.plot(train_losses, label='Training Loss')
plt.plot(val_losses, label='Validation Loss')
plt.title("Loss Curves across Epochs")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()

# 5. Report Evaluation Metrics (Requirement 6d)
model.eval()
with torch.no_grad():
    # Sigmoid applied because BCEWithLogitsLoss expects raw logits 
    y_prob_nn = torch.sigmoid(model(X_val_tensor.to(device))).cpu().numpy()
    y_pred_nn = (y_prob_nn > 0.5).astype(int)

# Use your previously defined classification metrics function
print_classification_metrics("Neural Network", y_val_2, y_pred_nn, y_prob_nn)
```
